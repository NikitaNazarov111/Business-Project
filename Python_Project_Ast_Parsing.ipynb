```python
import requests
from bs4 import BeautifulSoup
from requests.compat import urljoin, quote_plus, urlparse, unquote
import pandas as pd
```

С этого сайта мне надо запарсить 3 блока книг:

1) Фэнтези

2) Фантастика и фэнтези для детей и подростков

3) Фантастика

Сначала запарсю раздел "фэнтези":

```python
titles = []
authors = []
prices = []
for i in range(1, 86):
  url = f'https://ast.ru/cat/khudozhestvennaya-literatura/fentezi/?META.BOOK_PRICE_AVAILABLE=true&PAGEN_1={i}'
  response = requests.get(url)
  soup = BeautifulSoup(response.text, 'html.parser')
  books_sections = soup.find_all('div', class_='card__bottom card__bottom--offset-top card__bottom--fix-height')

  for section in books_sections:
    book_href = section.find('a')
    book_link = urljoin(url, book_href.get('href'))

    response_new = requests.get(book_link)
    soup_new = BeautifulSoup(response_new.text, 'html.parser')
    if soup_new.find('div', class_='price-block__price'):
      price_block = soup_new.find('div', class_='price-block__price')
      if price_block.find('span'):
        price = int(price_block.find('span').text.strip())
      else:
        price = 0
    else:
      price = 0

    if price == 0: #не буду парсить те книги, у которых нет цены
      continue
    else:
      title = section.find('a').text.strip()
      author = section.find('p', class_='card__text text-clip text-clip--small-xs text-clip--transparent-gray').text.strip()
      titles.append(title)
      authors.append(author)
      prices.append(price)
```

```python
df_ast_fantasy = pd.DataFrame({'Название' : titles, 'Автор' : authors, 'Цена' : prices})
df_ast_fantasy
```

```python
df_ast_fantasy.to_csv('ast_books_data_fantasy.csv', index=False, encoding='utf-8')
```
