```python
df_Chitay_Gorod_fantasy = pd.read_csv('/content/books_Fantasy_Chitay_Gorod.csv')
```

```python
df_Chitay_Gorod_kids = pd.read_csv('/content/books_kids_Chitay_Gorod.csv')
```

```python
df_Chitay_Gorod_teens = pd.read_csv('/content/books_teens_Chitay_Gorod.csv')
```

```python
df_Chitay_Gorod = pd.concat([df_Chitay_Gorod_fantasy, df_Chitay_Gorod_kids, df_Chitay_Gorod_teens], ignore_index=True)
df_Chitay_Gorod
```

```python
df_kids_labirint = pd.read_csv('/content/books_kids_lab.csv')
df_kids_labirint.columns=['Book_Title', 'Author', 'Publisher', 'Rating', 'Price(in Rubles)']
```

```python
df_Chitay_Gorod_copy = df_Chitay_Gorod.drop(columns=['Age_Restriction'])
df_merged_Lab_and_Chitay_Gorod = pd.concat([df_Chitay_Gorod_copy, df_kids_labirint], ignore_index=True)
df_merged_Lab_and_Chitay_Gorod
```

```python
from plotly.subplots import make_subplots
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
```

```python
fig = px.histogram(df_merged_Lab_and_Chitay_Gorod, x="Price(in Rubles)", color="Rating",
                  title="Цена = Качество?",
                  labels={"Price(in Rubles)": "Цена (руб.)"},
                  nbins=20)

fig.update_layout(
    title_x=0.5,
    title_font_size=20,
    yaxis_title="Количество книг",
    legend_title="Рейтинги"
)
fig.show()
```

Мы видим, что в диапазоне до 800 рублей сосредоточено одно из самых больших количеств книг, причем преобладают книги с рейтингами выше 4.3, что указывает на то, что даже книги с небольшими ценами имеют высокое качество написания. Если посмотреть на график стоимостей книг от издательств, то мы увидим, что в основном у них издаются книги из низкого ценового диапазона, что подводит нас к первой стратегии: а именно заказ у издательств недорогих книг, которые ялвляются наиболее высоко оцененными читателями. Так мы сможем захватить и тех, кто любит каяественную литературу и тех, у кого низкий бюджет на покупку книг.

Также можно заметить, что число книг заметно снижается с ростом цены.
В этом диапазоне преобладают книги с высокими рейтингами (в среднем выше чем 4.5)
Это подтверждает корреляцию между высокой ценой и высоким качеством, ну или высоким спросом на эксклюзивные издания. А потому имеет смысл закупать у издательств не только дешевые книги, но и эксклюзивные издательства, поскольку они высоко оцениваются читателями.

Но чем мы отличаемся от конкурентов? Мы будем создавать особенные зоны вокруг экслюзивных книг - например, у нас будет коллекционная комната книг, охраняемая двумя эльфами стражами. Внутри также будут эльфы-актеры, которые будут хорошо знать содержания эксплюзивных книг и рекомендовать их покупателям. Но мы не будем игнорировать и книги с низким ценовым сегментом - для них также будут созданы уникальные декорации, которых нет ни в каких других магазинах. Одной из идей является создание еженедельных ярмарок продажи книг с низкими ценами. У нас будут стоять тележки с книгами и рядом с ними эльфы-актеры, которые ознакомились с содержанием этих книг и тоже могут рассказывать о них

```python
fig = px.scatter(df_merged_Lab_and_Chitay_Gorod, x="Price(in Rubles)", y="Rating", color="Publisher",
                 hover_data=["Book_Title", "Author"],
                 title="Зависимость рейтинга от цены с разбивкой по издательствам")
fig.update_layout(
    title_x=0.5,
    title_font_size=20,
    yaxis_title="Рейтинг книг",
    xaxis_title="Цена книг",
    legend_title="Издательство"
)

fig.show()
```

```python
df_Chitay_Gorod_copy_new = df_Chitay_Gorod_copy.drop(columns=['Rating'])
df_MDK_copy = df_MDK.drop(columns=['Age_Category', 'Book_Genre'])
df_kids_labirint_copy = df_kids_labirint.drop(columns=["Rating"])
```

```python
df_merged_all = pd.concat([df_Chitay_Gorod_copy_new, df_MDK_copy, df_kids_labirint_copy], ignore_index=True)
df_merged_all.to_csv('merged_df.csv')
```

```python
publisher_avg = df_merged_all.groupby("Publisher")["Price(in Rubles)"].mean().sort_values(ascending=False).reset_index()
fig = px.bar(publisher_avg, x="Publisher", y="Price(in Rubles)",
             title="Средняя цена книг по издательствам",
             labels={"Price(in Rubles)": "Средняя цена (руб.)", "Publisher": "Издательство"})

fig.update_layout(
    title_x=0.5,
    title_font_size=20,
)

fig.show()
```

```python
fig = px.histogram(df_merged_all, x="Price(in Rubles)", color="Publisher",
                  title="Распределение книг по ценам",
                  labels={"Price(in Rubles)": "Цена (руб.)"},
                  nbins=30)

fig.update_layout(
    title_x=0.5,
    title_font_size=20,
    yaxis_title="Количество книг",
    legend_title="Издательства"
)
fig.show()
```

Как мы видим, книги от издательств Эксмо и АСТ, которые являются одними из тех изданий, с которыми мы будем сотрудничать, имеют в среднем низкие цены, а низкие цены - это привлечение покупателей, потому что большинство покупателей чувствительны к стоимости, особенно в сегментах детской литературы, фэнтези и молодёжных изданий, потому что их много. Благодаря этому у нас будут низкие закупочные цены и высокая рентабельность.

Анализ средних цен магазинов по издательствам позволит нам провести правильную ценовую политику. Например, мы сделаем средние цены ниже. Также вы узнаете, что мы будем продавать не только дешевые книги от издательств, но и дорогие, которых скорее всего нет у кокнкуернтов, что добавит нашему магазину эксклюзивности.

Что по особым предложениям и дизайну:

Мы разместим книги Эксмо и АСТ в зонах, где большая проходимость (рядом с эльфийским деревом, входами).

Также в начале мы будем снижать цены на книги-бестселлеры Эксмо/АСТ - будем продавать на 10–15% ниже среднерыночных, чтобы привлечь аудиторию.

Помимо этого, мы создадим отдельные зоны с этими издательствами, поскольку они являются нашими основными поставщиками. Дополнительная реклама издательств поможет нам снизить закупочные цены, поскольку так мы повышаем популярность этих изданий.

К примеру, можно будет добавить «Лес эльфов» — детская литература и фэнтези с актерами в костюмах, читающими отрывки из книг.

Наконец, поскольку Эксмо и АСТ будут нашими ключевыми поставщиками, мы договримся с ними о выпуске эксклюзивных версий книг с уникальным оформлением в виде эльфийского мира. Это будет что-то по типу линейки книг, которых нет нигде.
