```python
import requests
from bs4 import BeautifulSoup
from requests.compat import urljoin, quote_plus, urlparse, unquote
import pandas as pd
```

Парсю 1 источник (сайт издательства "Самокат"):

Сначала спарсю книги из раздела "младшим школьникам" (подраздел "фантастика для детей")

```python
url = 'https://samokatbook.ru/catalog/detskaya-hudozhestvennaya-literatura/fantastika/'

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
books_sections = soup.find_all('div', class_='main-goods_item main-goods_item__display')
```

```python
titles = []
prices = []
authors = []
for section in books_sections:
  unavailable = section.find('div', class_='activity-goods-cart')
  if unavailable and 'нет в наличии' in unavailable.text.strip():
    continue
  title = section.find('a', class_='main-goods_name no-color').text.strip()
  titles.append(title)

  price_block = section.find('div', class_='main-goods-price')
  for span in price_block.find_all('span'):
    span.decompose()
  price = price_block.text.strip()
  only_numbers_price = [x for x in price if x.isdigit()]
  price = int(''.join(only_numbers_price))
  prices.append(price)

  author = section.find('div', class_='main-goods-author').text.strip()
  authors.append(author)
```

Теперь спарсю еще книги из раздела "подросткам" (подраздел "фантастика для подростков"). Там есть пересечения, поэтому если в titles уже есть такое название, то буду переходить к следующей книге.

```python
url = 'https://samokatbook.ru/catalog/knigi-podrostki/fentezi-fantastika-podrostok/'

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
books_sections = soup.find_all('div', class_='main-goods_item main-goods_item__display')
```

```python
for section in books_sections:
  unavailable = section.find('div', class_='activity-goods-cart')
  if unavailable and 'нет в наличии' in unavailable.text.strip():
    continue
  title = section.find('a', class_='main-goods_name no-color').text.strip()
  if title in titles:
    continue
  else:
    titles.append(title)

    price_block = section.find('div', class_='main-goods-price')
    for span in price_block.find_all('span'):
      span.decompose()
    price = price_block.text.strip()
    only_numbers_price = [x for x in price if x.isdigit()]
    price = int(''.join(only_numbers_price))
    prices.append(price)

    author = section.find('div', class_='main-goods-author').text.strip()
    authors.append(author)
```

```python
df_samokat = pd.DataFrame({'Название' : titles, 'Автор' : authors, 'Цена' : prices})
```

```python
df_samokat.to_csv('samokat_books_data.csv', index=False, encoding='utf-8')
```
